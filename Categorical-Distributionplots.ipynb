## Categorical Distributionplots

1.Boxplot  
2.Violinplot  

X-axis contains the Categorical data  
Y-axis contains the Numerical data

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('fivethirtyeight')
```

```python
data= sns.load_dataset('tips')
```

```python
data.head()
```

```python
sns.catplot(x='tip',kind='box',data=data)
```

```python
sns.catplot(x='time',y='tip',kind='box',data=data)
```

```python
sns.catplot(x='time',y='tip',kind='box',data=data,hue='sex')
```

Violin plot are considered as the combination of Boxplot with a kde.

```python
sns.catplot(x='tip',kind='violin',data=data)
```

```python
sns.catplot(x='time',y='tip',kind='violin',data=data)
```

```python
sns.catplot(x='time',y='tip',kind='violin',data=data,hue='sex')
```

```python
sns.catplot(x='day',y='tip',kind='violin',data=data,hue='sex',split=True)
```
