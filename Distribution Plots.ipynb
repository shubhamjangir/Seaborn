## Distribution Plots
1.Jointplots  
2.Pairplots  
3.Displots     

X-axis contains the Numerical data  
Y-axis contains the Numerical data

```python
import seaborn as sns
```

```python
data = sns.load_dataset('tips')
```

```python
data.head()
```

```python
sns.jointplot(x='tip',y='total_bill',data=data)
```

```python
sns.jointplot(x='tip',y='total_bill',data=data,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=data,kind='kde')
```

```python
sns.jointplot(x='tip',y='total_bill',data=data,kind='reg')
```

```python
data=sns.load_dataset('iris')
```

```python
sns.pairplot(data,hue='species')
```

```python
data.head()
```

```python
sns.distplot(data['sepal_length'])
```

```python
sns.distplot(data['sepal_length'],kde=False)
```

```python
sns.distplot(data['sepal_length'],hist=False)
```

```python
sns.distplot(data['sepal_length'],hist=False,kde=False,rug=True)
```
