## Relational/Statistical Plot

Relational plots are used to perform multivariate analysis

1.Scatter Plot  
2.Line Plot  
3.FacetGrids   

Figure level- relplot()   
Axes level- scatterplot()

### Scatter Plot

This is a plot between 2 numerical entities and 3,4th category parameter can be given to get better view and understanding.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=sns.load_dataset('tips')
```

```python
data.head()
```

### total_bill vs tip

```python
sns.relplot(x='total_bill',y='tip',kind='scatter',data=data)
```

### hue

```python
sns.relplot(x='total_bill',y='tip',kind='scatter',data=data,hue='smoker')
```

### style

```python
sns.relplot(x='total_bill',y='tip',kind='scatter',data=data,hue='smoker',style='sex')
```

### size

```python
sns.relplot(x='total_bill',y='tip',kind='scatter',data=data,hue='smoker',style='sex',size='size',sizes=(100,200))
```

### col

```python
sns.relplot(x='total_bill',y='tip',kind='scatter',data=data,hue='smoker',style='sex',col='time')
```

### row

```python
sns.relplot(x='total_bill',y='tip',kind='scatter',data=data,hue='smoker',style='sex',col='time',row='day')
```

### col_wrap

```python
sns.relplot(x='total_bill',y='tip',kind='scatter',data=data,hue='smoker',style='sex',col='day',col_wrap=2)
```

Same thing can be done use scatter plot instead of relplot and using scatter as kind 

```python
sns.scatterplot(x='total_bill',y='tip',data=data)
```
