## Regression Plots

1.Regplot   
2.lmplot

X-axis contains the Numerical data  
Y-axis contains the Numerical data

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
data=sns.load_dataset('tips')
```

```python
data.head()
```

```python
sns.regplot(x='total_bill',y='tip',data=data)
```

```python
sns.regplot(x='total_bill',y='tip',data=data,color='r',marker='+')
```

```python
sns.regplot(x='total_bill',y='size',data=data,color='r',marker='+',x_jitter=.1)
```

```python
sns.regplot(x='total_bill',y='size',data=data,color='r',marker='+',x_jitter=.1,x_estimator=np.mean)
```

```python
sns.lmplot(x='total_bill',y='tip',data=data,hue='sex',col='smoker',row='time')
```

```python
sns.lmplot(x='total_bill',y='tip',data=data,hue='sex',col='day',col_wrap=2)
```
