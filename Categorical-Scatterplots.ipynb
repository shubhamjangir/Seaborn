## Categorical Plots

Figure level-catplot()  
Axis level- subsequent name of the plots

## Categorical Scatterplots
1.Striplot   
2.Swarmplot

X-axis contains the Categorical data  
Y-axis contains the Numerical data

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
```

```python
data=sns.load_dataset('tips')
```

```python
data.head()
```

```python
sns.catplot(x='day',y='tip',kind='strip',data=data)
```

```python
#Jitter- Random Noise
sns.catplot(x='day',y='tip',kind='strip',data=data,jitter=False)
```

```python
sns.catplot(x='day',y='tip',kind='swarm',data=data)
```

```python
sns.catplot(x='day',y='tip',kind='swarm',data=data,hue='sex')
```
