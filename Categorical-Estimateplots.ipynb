## Categorical Estimateplots

1.Barplot   
2.Countplot

X-axis contains the Categorical data   
Y-axis contains the Numerical data

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
```

```python
data=sns.load_dataset('tips')
```

```python
data.head()
```

```python
sns.catplot(x='smoker',y='total_bill',data=data,kind='bar')
```

```python
sns.catplot(x='smoker',y='total_bill',data=data,kind='bar',hue='sex')
```

```python
sns.catplot(x='smoker',y='total_bill',data=data,kind='bar',hue='sex',estimator=np.median)
```

```python
sns.catplot(x='smoker',y='total_bill',data=data,kind='bar',hue='sex',estimator=np.std)
```

```python
sns.catplot(x='smoker',data=data,kind='count')
```

```python
sns.catplot(x='sex',data=data,kind='count')
```
